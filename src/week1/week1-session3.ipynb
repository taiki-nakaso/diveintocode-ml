## 栗まんじゅう問題
漫画ドラえもんには「バイバイン」という道具が登場する、2のn乗に関する有名なエピソードがあります。


「バイバイン」は薬品で、これを振掛けるとその物体は5分ごとに倍に増えるという働きがあります。作中では栗まんじゅうを増やしますが、食べきれなくなったためロケットを使い宇宙に捨てました。


ここで度々話題になるのが、栗まんじゅうはその後宇宙でどのように増えていくのかについてです。様々な議論がされていますが、今回は単純に栗まんじゅうの体積がどう増えるかを計算してみましょう。



# 問題
## 栗まんじゅうが太陽系を覆う日
1個の栗まんじゅうにバイバインを振りかけた場合、太陽系が栗まんじゅうで覆われるのは何分後でしょうか。これを計算するプログラムを作成してください。


栗まんじゅうや太陽系の体積を調査した上で仮定して計算してください。栗まんじゅうの隙間や重力などは考えず、問題を単純化して構いません。


ボールにバイバインを振りかけた時、東京ドームが一杯になる時間など他の状況にも適用できるような汎用的なプログラムを作成してください。また、グラフを表示するなど分かりやすいノートブックを作成してください。

栗まんじゅうの体積 = 100(cm^3) = 1e-4(m^3)

太陽系の体積 = (4/3) \* math.pi \* (45*10e+11)^3

とする。

なお、簡単のために栗まんじゅうの体積は100cm^3に、太陽系の体積については太陽-海王星間の距離を半径とする完全な球体であると仮定する。

```python
"""
@param covering_vol バイバインかけられる方の物体の体積
@param covered_vol 増えた物体に覆われる空間の体積
@return time_list 覆いきるまでの時間のリスト(5分刻み)
@return volume_list 覆いきるまでの合計体積のリスト
"""
def baibain(covering_vol=0, covered_vol=0):
    # 返り値用リスト　時間と値を返す
    time_list = [0]
    volume_list = [covering_vol]
    # カウンタ, temp変数
    double_count = 0
    tmp_vol = covering_vol

    while(tmp_vol < covered_vol):
        tmp_vol *= 2
        double_count += 1
        time_list.append(double_count*5)
        volume_list.append(tmp_vol)
    
    return time_list, volume_list
```

```python
import matplotlib.pyplot as plt
import math

KURI_V = 1e-4 # 栗まんじゅう
SOLSYS_V = (4/3) * math.pi * (45*10e+11)**3 # 太陽系

times, volumes = baibain(covering_vol=KURI_V, covered_vol=SOLSYS_V)

print("栗まんじゅうが太陽系を覆う時間: {}分".format(times[-1]))

%matplotlib inline
plt.title('volume increasing')
plt.xlabel('time[min]')
plt.ylabel('volume[m^3]')
plt.plot(times, volumes, label='real volume')
plt.plot(times, [SOLSYS_V]*len(volumes), '--r', label='solorsystem volume')
plt.legend()
plt.show()
```

```python
# 野球ボールが東京ドームを覆う時間を考えてみる

BASEBALL_V = (4/3) * math.pi * (74.84e-3/2)**3 # 硬球
TOKYODOME_V = 124e+4 # 東京ドーム

times, volumes = baibain(covering_vol=BASEBALL_V, covered_vol=TOKYODOME_V)

print("野球ボールが東京ドームを覆う時間: {}分".format(times[-1]))

%matplotlib inline
plt.title('volume increasing')
plt.xlabel('time[min]')
plt.ylabel('volume[m^3]')
plt.plot(times, volumes, label='real volume')
plt.plot(times, [TOKYODOME_V]*len(volumes), '--r', label='Tokyo Dome volume')
plt.legend()
plt.show()
```

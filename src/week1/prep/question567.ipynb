# 問題5
紙を折る過程の値をリストに保存する。
0回-43回までの44点がリストに格納されることを確認する。

```python
THICKNESS = 8e-5
# 値格納用のリスト
thickness_list = []
```

```python
# 初期値格納
folded_thickness = THICKNESS
thickness_list.append(folded_thickness)
```

```python
# 折る
for i in range(43):
    folded_thickness *= 2
    thickness_list.append(folded_thickness)
```

```python
# 出力
print(thickness_list)
print("len(thickness_list): {}".format(len(thickness_list)))
```

# 問題6
問題5のデータを利用し、 `Matplotlib` を用いてグラフを描画する。

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("thickness of folded paper")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(thickness_list)
plt.show()
```

# 問題7
先ほどのグラフをカスタマイズしたものを *3種類以上* 作成する。

今回は3種類のグラフを追加で作成する。

```python
%matplotlib inline
# graph 1
plt.title("thickness of folded paper[1]")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(thickness_list, 'rx')
plt.show()

%matplotlib inline
# graph 2
plt.title("thickness of folded paper[2]")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.yscale('log')
plt.plot(thickness_list, linewidth=5)
plt.show()

%matplotlib inline
# graph 3
plt.title("thickness of folded paper[3]")
plt.xlabel("number of folds")
plt.ylabel("thickness[m]")
plt.plot(thickness_list, '--g', label='thickness')
plt.legend()
plt.grid(b=True, which='major', axis='y')
plt.show()
```
